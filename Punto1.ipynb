```python
%pylab inline
import numpy as np
import scipy.optimize as scipy
import matplotlib.pyplot as plt
```

```python
#Matriz de covarianza
sigma = ((1/5.,0),(0,1/5.))
#centros de las distribuciones
cen_blue = [1 , 0]
cen_oran = [0 , 1]
```

```python
#Genera los centroides de los puntos azules y naranjas: Puntos, color
def centroides(num_centroides, cen_blue, cen_oran, sigma):
    pois_blue = np.zeros((num_centroides, 2))
    pois_oran = np.zeros((num_centroides, 2))
    color = np.zeros((num_centroides*2))
    for i in range(num_centroides):
        pois_blue[i, 0], pois_blue[i,1]= np.random.multivariate_normal((cen_blue[0],cen_blue[1]), sigma)
        pois_oran[i, 0], pois_oran[i,1]= np.random.multivariate_normal((cen_oran[0],cen_oran[1]), sigma)
        #color azul la primera mitad
        color[i] = 1
    puntos = np.concatenate((pois_blue, pois_oran))
    return puntos, color
```

```python
#graficar centroides
num_cent = 10
puntos_cent, color_cent = centroides(num_cent, cen_blue, cen_oran, sigma)
plt.title("Centroides puntos azules y rojos")
plt.scatter(puntos_cent[:, 0], puntos_cent[:, 1], marker='o', c = color_cent, s=100)
plt.xlabel('x')
plt.ylabel('y')
```

```python
#Generador observaciones
def generador_observaciones(cents, sigma, numero_obs):
    puntos = np.zeros((numero_obs,2))
    j = 0
    for i in range(numero_obs):
        if j == len(cents)-1:
            j = 0
        puntos[i,0], puntos[i,1] = numpy.random.multivariate_normal((cents[j,0],cents[j,1]),sigma) 
        j += 1
    return puntos
```

```python
def poblacion(number, puntos_cent, color_cent):
    blue = generador_observaciones(puntos_cent[color_cent == 1], sigma, number)
    oran = generador_observaciones(puntos_cent[color_cent == 0], sigma, number)
    puntos = np.concatenate((blue, oran))
    color = np.zeros((number*2))
    #primera mitad puntos azules
    for j in range (number):
        color[j] = 1
    return puntos, color
```

```python
#Muestra de entrenamiento y de test
pts_trai, clr_trai = poblacion(100, puntos_cent, color_cent)
pts_test, clr_test = poblacion(5000, puntos_cent, color_cent)
```

```python
#Grafica puntos entrenamiento
plt.title(u"Población puntos de entrenamiento")
plt.scatter(pts_trai[:, 0], pts_trai[:, 1], marker='o', c = clr_trai, s=100)
plt.xlabel('x')
plt.ylabel('y')
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(pts_trai , clr_trai)
beta = reg.coef_
pendiente = - beta[0]/beta[1]
inter = reg.intercept_

x_1 = linspace(pts_trai[:,0].min(),pts_trai[:,0].max(),100)
x_2 = pendiente * x_1 - (inter - 0.5)/ beta[1]
```

```python
#Grafica
plt.title('Modelo lineal')
plt.scatter(pts_trai[:, 0], pts_trai[:, 1], marker='o', c= clr_trai, s=100)
plt.plot(x_1, x_2, label= 'linear')
plt.xlabel('x')
plt.ylabel('y')
```

```python
def encontrar_error(reg_linear, puntos, color):
    Y_predict = reg_linear.predict(puntos)
    puntos_mal = 0
    for i in range(len(puntos)):
        #malos azules
        if(color[i] == 1 and Y_predict[i] < 0.5):
            puntos_mal += 1
        elif(color[i] == 0 and Y_predict[i] > 0.5):
            puntos_mal += 1
    effec = puntos_mal/float(len(Y_predict))
    return effec
```

```python
error_trai = (1 - encontrar_error(reg, pts_trai, clr_trai))*100
error_test = (1 - encontrar_error(reg, pts_test, clr_test))*100
print "La efectividad del modelo lineal en la muestra de entrenamiento es del " + str(error_trai) + "%"
print "La efectividad del modelo lineal en la muestra de prueba es del " + str(error_test) + "%"
```

```python
#Error en funcion de cantidad de vecinos, grados libertad = cantidad puntos/cantidad vecinos
def error_vecinos(mini, maxi, puntos, color):
    vecinos = np.logspace(np.log10(mini), np.log10(maxi), 15)
    grados_lib = []
    errores = []
    for i in range(len(vecinos)):
        reg_veci = neighbors.KNeighborsClassifier(int(vecinos[i]))
        reg_veci.fit(puntos, color)
        grados_lib.append(len(puntos)/double(vecinos[i]))
        error = encontrar_error(reg_veci, puntos, color)
        errores.append(error)          
    return grados_lib, errores
```

```python
#Define la cantidad de vecinos para cada muestra de manera que ambos tengan de 1 a 200 grados de libertad
min_trai, max_trai = 1, len(pts_trai)/2
min_test, max_test = 50, len(pts_test)/2                          
grados_trai, error_trai = error_vecinos(min_trai, max_trai, pts_trai, clr_trai)
grados_test, error_test = error_vecinos(min_test, max_test, pts_test, clr_test)
```

```python
plt.title('Curvas de error')
plt.semilogx(grados_trai, error_trai,"o--", label = 'Entrenamiento', color = 'r')
plt.semilogx(grados_test, error_test,"o--", label = 'Test')
plt.xlabel('Grados de libertad - N/k')
plt.ylabel('Error')
plt.legend()
```
