```python
%pylab inline
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
#Hace una marcha aleatoria en 1 dimensión de n pasos, devuelve la posicion final
def marcha_aleatoria(n):
    posicion = []
    posicion_actual = 0
    for i in range(n):
        m = random.randint(0,1)
        if (m == 0):
            posicion_actual += 1
        else:
            posicion_actual -= 1
        posicion.append(posicion_actual)
    return posicion[-1]
```

La siguiente función calcula los p-values usando la fórmula:
$p_{value} = \frac{X_{exp} - H}{\sigma } $
Donde $X_{exp}$ es el promedio del valor en cada experimento, $H$ es el valor esperado para el promedio y 
$\sigma$ es la varianza de las mediciones.

```python
#Realiza m marchas aleatorias de n pasos devuelve el p-value de cada experimento
def experimento(m ,n):
    posicion_final = []
    n_pasos = 10
    suma = 0
    for i in range(m):
        posicion_final.append(marcha_aleatoria(n_pasos))
        suma += posicion_final[i]**2
    promedio = np.mean(posicion_final)
    sigma = np.sqrt(suma)
    p_value = abs(promedio)/sigma
    return p_value
```

```python
#Realiza w veces cada experimento (m marchas aleatorias de m pasos) y devuelve los p-values
def varios_experimentos(w, m, n):
    p_values = []
    for i in range (w):
        p_values.append(experimento(m,n))    
    return p_values
```

```python
m = 1000
n = 100
w = 1000
p_values = varios_experimentos(w,m,n)
```

```python
plt.title(u'Distribución de los p-values')
plt.hist(p_values, 15)
plt.xlabel('Valor de p-value')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Calcula las propiedades de los p_values:
def propiedades(max_exp, m, n):
    promedios = []
    minimos = []
    maximos = []
    for i in range(max_exp):
        p_values = varios_experimentos(i+1,m,n)
        promedios.append(np.mean(p_values))
        minimos.append(np.min(p_values))
        maximos.append(np.max(p_values))
    return minimos, promedios, maximos
```

```python
n = 100
m = 200
max_exp = 30
minimos, promedios, maximos = propiedades(max_exp, m , n)
numero_exp = np.linspace(0, max_exp, max_exp)
title("Propiedades distribucion de p-values")
plt.plot(numero_exp, maximos, "o--", label = "Maximos", color = 'b')
plt.plot(numero_exp, promedios, "o--", label = "Promedio", color = 'r')
plt.plot(numero_exp, minimos, "o--", label = "Minimos", color = 'g')
plt.xlabel(u'Número de experimentos')
plt.ylabel('Valor p-value')
plt.legend(loc='upper right')
plt.show()
```

```python
def graficar_propiedades(n,m, max_exp):
    minimos, promedios, maximos = propiedades(max_exp, m , n)
    numero_exp = np.linspace(0, max_exp, max_exp)
    title('Minimos n= ' + str(n)) 
    plt.scatter(numero_exp, minimos, label = "Minimos")    
```

```python
#Realiza diferentes graficas del valor minimo de p varia el numero de pasos, en promedio se observa una
#distribución parecida a la mostrada en el papper
plt.figure(figsize=(15,15))
n = 100
m = 200
max_exp = 30
for j in range(1,17):
    plt.subplot(4,4,j)
    graficar_propiedades(n, m, max_exp)
    n += 10
plt.show()  
# Con mayores valores de m da mejor la grafica de los minimos p_values m= 200
```
